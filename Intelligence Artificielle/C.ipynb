```python
import pandas as pd
from joblib import dump
from joblib import load
# rtr = load('modele_T.joblib')
dts = pd.read_parquet("Importation 2021.parquet", columns=['Flux','Trimestre','Code Departement','Code Region','A129','CPF4','Code Pays','Valeur','Masse','Pays','UE','Taux Douane','NC8','CPF6','A17','A38'])
Faux = pd.read_parquet("Faux.parquet")
dts.dropna(inplace=True)
dts = dts.head(1000000)
```

```python
dts['Flux'] = (dts['Flux'] == 'I')
dts['Code Departement'] = dts['Code Departement'].cat.codes
dts['Code Region'] = dts['Code Region'].cat.codes
dts['A129'] = dts['A129'].cat.codes
dts['CPF4'] = dts['CPF4'].cat.codes
dts['Code Pays'] = dts['Code Pays'].cat.codes
dts['Pays'] = dts['Pays'].cat.codes
dts['Taux Douane'] = dts['Taux Douane'].cat.codes
dts['NC8'] = dts['NC8'].cat.codes
dts['CPF6'] = dts['CPF6'].cat.codes
dts['A17'] = dts['A17'].cat.codes
dts['A38'] = dts['A38'].cat.codes

print(dts)
dts.info()
```

```python
dts['Prix'] = dts['Valeur'] - (dts['Valeur'] * dts['Taux Douane'].astype(float) / 100)
print(dts)
dts.info()
```

```python
dts['Etat'] = 1
Faux['Etat'] = 0
dts = pd.concat([dts,Faux])
dts["Taux Revient"] = ( dts['Valeur'] + dts['Prix'] ) / dts['Valeur']
print(dts)
```

```python
dts.dropna(inplace=True)
print(dts)
dts.info()
```
